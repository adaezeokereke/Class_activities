```python
#simple quiz game

questions = ("What is the most spoken language in the world?: ",
             "Which is the smallest country in the world?: ",
            "Which is the longest river in the world?: ",
            "Which element has the chemical symbol \'O'?: ",
            "How many bones are there in the human body?: ",
            "What is the capital city of USA?: ",
            "Which planet is known as the red planet? ",
            "What is the capital city of France? ",
            "Which is the largest planet in the solar system? ",
            "Which planet is known as ringed planet? "
)

options = (("A: Mandarin", "B: English", "C: Spanish", "D: German"),
           ("A: Vatican City", "B: Monaco", "C: San Marino", "D: Tokyo"),
           ("A: Amazon", "B: Nile", "C: Yangtze", "D: Niger"),
           ("A: Gold", "B: Oxygen", "C: Silver", "D: Nitrogen"),
           ("A: 206", "B: 195", "C: 213", "D: 204"),
           ("A: New York", "B: Washington DC", "C: Los Angeles", "D: Florida"),
           ("A: Venus", "B: Mars", "C: Jupiter", "D: Saturn"),
           ("A: Rome", "B: Berlin", "C: Paris", "D: Japan"),
           ("A: Neptune", "B: Saturn", "C: Jupiter", "D: Mars"),
           ("A: Mars", "B: Saturn", "C: Venus", "D: Jupiter"))

answers = ("B", "A", "B", "B", "A", "B", "B", "C", "C", "B")

possible_answers =[]

score = 0

question_num = 0

for question in questions:
    print(question)
    for option in options[question_num]:
        print(option)

    guess = input("Enter (A, B, C, D): ").upper()
    possible_answers.append(guess)
    if guess == answers[question_num]:
        score += 1
        print('CORRECT!')
    else:
        print("INCORRECT!")
        print(f"{answers[question_num]} is the correct answer")

    question_num += 1


print("     Results     ")

print('answers: ', end="")
for answer in answers:
    print(answer, end=' ')
print()

print("guesses: ", end= " ")
for guess in answers:
    print(answer, end= " ")
print()

score =int(score / len(questions) * 100)
print(f"Your score is: {score}% ")
```
