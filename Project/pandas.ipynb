```python
#creating pandas dataframe from lists using the zip method
#list 1
name=['Ayo','Imran','Chucks','Judith']

#list 2
Age = [25,30,26,22]

#list 3
state= ['Abuja','Sokoto','Kano','Abia']

#merging the lists using the zip method

new_list=list(zip(name,Age,state))
new_list
```

```python
new=pd.DataFrame(new_list)
new
```

```python
#pandas DataFrame from lists using zip
#list 1
Name=['Ayo','Imran','Chucks','Judith']

#list 2
Age = [25,30,26,22]

#get the list of tuples from two lists and merge using zip()
list_of_tuples=list(zip(Name,Age))


list_of_tuples
```

```python
#converting lists of tuples into pandas DataFrame
df=pd.DataFrame(list_of_tuples,columns = ['Name','Age'])

list_of_tuples
```
